# Определение целевых метрик для Lamoda (в 2011 году)

Я не уверен что я правильно понимаю идею сегментации KPI, я отталкивался от той идеи что низкоуровневые KPI - это работа тех метрик которые большие боссы с чистой совестью делегируют молодым работягам, а высокоуровневые это более организационные абстрактные бизнесовые метрики.

По крайней мере я надеюсь в 2011 году тоже далеко не все понимали разницу и воспринимали это примерно как я

## Высокоуровневые 

### Маркетинговые 

* Лояльность пользователей
* Суммарная стоимость клиента
* Количество пользователей из разных источников трафика

### Продуктовые 

* Выручка
* Прибыль
* Маржинальность товаров
* Расходы на оффлайн составляющую (логистика)
* Суммарная прибыль от клиента

## Низкоуровневые 

### Маркетинговые 

* Количество переходов на сайт
* Количество регистраций на сайте
* Количество отказов (подразумеваются люди которые покинули сайт на первой странице)
* Глубина просмотра / Время нахождения на сайте

### Продуктовые 

* Средний чек
* Количество возвратов
* Время с момента входа на сайт, до совершения целевого действия
* Отношение возвратов к количеству покупок
* Время от совершения покупки до доставки покупки

#  Как, кем и с какой периодичностью должны мониториться KPI? 


В моих глазах сегментация метрик подразумевает собой следующее подразделение:

1) Высокоуровневые метрики мониторятся высокоуровневым персоналом (топ-менеджерами, партнёрами, директорами) и мониторятся довольно редко. Раз в месяц или раз в квартал, так как изменение стратегии относительно работы с этими метриками кажется мне довольно сложной задачей, скажем, перераспределить расходы на диаметрально противоположные источники трафика организационно не так просто и довольно рискованно.

2) Низкоуровневые метрики, в свою очередь, делегируются более нижним ступеням менеджеров и аналитиков и мониторятся значительно чаще. Мне кажется, в принципе, любое организационное решение должно приниматься на основе метрик и их изменений, следовательно если условный менеджер Тамара будет ежедневно смотреть как её решения влияют на показатели метрик - будет идеально.
